```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
euro_cities = pd.read_csv('hoursspentincongestion.csv')
```

```python
euro_cities.head()
```

```python
euro_cities['congestion_by_density'] = (euro_cities['hours_spent_in_congestion'] / (euro_cities['density_km2'] /  1000)).map(lambda x: round(x, 1))
```

```python
euro_cities.congestion_by_density.median()
```

```python
euro_cities = euro_cities.sort_values(by='congestion_by_density', ascending = False)

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_style("whitegrid")
euro_congestion = sns.barplot(x=euro_cities['city'], y=euro_cities['congestion_by_density'], color='#637eb6')
euro_congestion.set_xticklabels(labels=euro_cities['city'],rotation=90)


plt.title('Congestion of Major European Cities')
plt.xlabel('City')
plt.ylabel('Hours in Congestion by Density (in 1000s)')
plt.show()
```
